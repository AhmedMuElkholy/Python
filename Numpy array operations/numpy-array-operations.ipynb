## Function 1 - np.empty

This function helps you to create numpy arrays without initialize the entry of given shapes and types, it's like np.zeros but it's quite faster

```python
# Example 1 - working
np.empty(2, dtype=int)
```

Created random (2,0) array of type int

```python
# Example 2 - working
np.empty((3,5), dtype=float, order = 'C')
```

You can define the size of the array

```python
# Example 3 - not working
np.empty(2, dtype='datetime64')
```

In this example I tried to make an numpy array of datetime and it gave error,

## Function 2 Copyto

this function copies values from np array to another one, the formula is

numpy.copyto(dst, src, casting='same_kind', where=True)

```python
# Function 1 working
gfg = np.array([1, 2, 3])
print(gfg)
geeks = [4, 5, 6]
print(geeks)      
np.copyto(gfg, geeks)
print(gfg)
```

In this example we can see that with the help of numpy.copyto() method we are making the copy of an elements in a destination array.

```python
# Function 2 working
gfg = np.array([[1, 2, 3], [4, 5, 6]])
geeks = [[4, 5, 6], [7, 8, 9]]
np.copyto(gfg, geeks)
print(gfg)
```

We can copy list of arrays as well

```python
#Function 3 not working
gfg = np.array([1, 2, 3, 4])
geeks = [4, 5, 6]
np.copyto(gfg, geeks)
```

Both arrays should have the same shape

## Function 3 - Np.equal

Compares whether the 2 arrays values are equal

```python
#Function 1 working
a = np.equal([1., 2.], [1., 3.])
print("Check to be Equal : \n", a, "\n")

b = np.equal([1, 2], [[1, 3],[1, 2]])
print("Check to be Equal : \n", b, "\n")
```

In the second example the check ran to compare values with the first array

```python
#Function 2 working
a = np.array([6+7j, 2])
b = np.array([1,2])

d = np.equal(a, b)
print("Comparing complex with int using .equal() : \n ", d)
```

 Comparing data-type using .equal() function

```python
#Function 3 not working
a = np.equal(['1.', 2.], [1., 3.])
print("Check to be Equal : \n", a, "\n")
```

The 2 arrays must have the same class

## Function 4 - numpy.sort()

Return a sorted copy of an array.



```python
#Function 1 working
a = np.array([[12, 15], [10, 1]])
arr1 = np.sort(a, axis = 0)        
print ("Along first axis : \n", arr1)        
 
```

sorts the array on the columns axis

```python
#Function 2 working
a = np.array([['a', 15], ['b', 1]])
arr1 = np.sort(a, axis = 0)        
print ("Along first axis : \n", arr1)        
 
```

It can sort strings as well

```python
#Function 3 not working
a = np.array([['a', 15], ['b', 1]])
arr1 = np.sort(a, axis = 2)        
print ("Along first axis : \n", arr1)        
 
```

Axis should be 1 or 0

## Function 5 - bitwise

There are functions to convert the elements into their binary representation and then apply operations on the bits.

```python
#Function 1 working
a=10
b=11
print(bin(a))
print(bin(b))
np.bitwise_and(a,b)
```
